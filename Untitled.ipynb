```python
from ics import Calendar

# Read an .ics file
with open("events.ics", "r") as f:
    calendar = Calendar(f.read())

# Loop through events
for event in calendar.events:
    print("Title:", event.name)
    print("Start:", event.begin)
    print("End:", event.end)
    print("Description:", event.description)
    print("Location:", event.location)
    print("-" * 40)
```

```python
import requests
from ics import Calendar

url = "https://doyle.cbsd.org/fs/calendar-manager/events.ics?calendar_ids=6"

response = requests.get(url)
response.raise_for_status()  # will throw an error if the request fails

calendar = Calendar(response.text)

# Loop through events
events = [[event.name, event.begin.strftime('%Y-%m-%d'), event.end.strftime('%Y-%m-%d'), event.description, event.location] for event in calendar.events]
events
```
